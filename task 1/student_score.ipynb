*   Predict the percentage of a student bases on the number of study hours
*   this is a simple linear regerssion task as it involves just 2 variables
*   you can use R, SAS, Python or any other tool
*   data can be found at http://bit.ly/w-data
*   what will be the predicted score if a student studies for 9.25 hours a day

```python
#importing the necessary modules
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
#reading the dataset
df = pd.read_csv('http://bit.ly/w-data')
df.head()
```

```python
#defining predictor and response variables
X = df[['Hours']]
Y = df['Scores']
```

```python
plt.scatter(X,Y)
```

```python
#splitting the dataset into training and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state = 0)
```

```python
#creating linear regression object
lm1 = LinearRegression()
```

```python
#training the model
lm1.fit(X_train,Y_train)
```

```python
#predicting the scores
yhat = lm1.predict(X_test)
yhat
```

```python
df = pd.DataFrame({'Actual':Y_test,'Predicted':yhat})
df
```

```python
# Plotting the regression line
line = lm1.coef_*X+ lm1.intercept_
# Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line);
plt.title('Hours vs Score')  
plt.xlabel('Hours')  
plt.ylabel('Scores')
plt.show()
```

```python
#checking the accuracy scores for training and testing
print('Test Score')
print(lm1.score(X_test, Y_test))
print('Training Score')
print(lm1.score(X_train, Y_train))
```

```python
#Checking the efficiency of model
mean_squ_error = mean_squared_error(Y_test, yhat[:9])
mean_abs_error = mean_absolute_error(Y_test, yhat[:9])
print("Mean Squred Error:",mean_squ_error)
print("Mean absolute Error:",mean_abs_error)
```

```python
#predicting score of the students who spend 9.25 hours a day
new_score = lm1.predict([[9.25]])
print('score of students who spend 9.25 hours a day studying: ',new_score[0])
```
